```python
import cv2

# Load video
video_path = "drone.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define output writer
out = cv2.VideoWriter('output_with_spray.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection on each frame
    results = model(frame, verbose=False)

    # Draw detections and check for "window" class
    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            label = model.names[cls_id]

            # If it detects 'window'
            if label.lower() == "window" and conf > 0.5:
                cv2.putText(frame, "Spray initiated on window!", (50, 50),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                print("Spray initiated on window!")  # also logs in console

    out.write(frame)
    cv2.imshow("Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import pyttsx3

# ===============================
# 1️⃣ Load Model
# ===============================
# Use your trained model (not YOLOv8n.pt)
model = YOLO("runs/detect/train2/weights/best.pt")  # <-- replace with your trained weights if custom (e.g., "runs/detect/train/weights/best.pt")

# ===============================
# 2️⃣ Initialize
# ===============================
engine = pyttsx3.init()
video_path = "drone.mp4"
cap = cv2.VideoCapture(video_path)

fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("output_with_spray.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (width, height))

spray_announced = False  # avoid repeating sound

# ===============================
# 3️⃣ Detection + Visualization Module
# ===============================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO on frame
    results = model(frame, verbose=False)

    # Plot default YOLO detections (just to confirm boxes appear)
    annotated_frame = results[0].plot()  

    window_detected = False

    # Manually check detections for “window”
    for box in results[0].boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        label = model.names[cls_id]

        if label.lower() == "window" and conf > 0.5:
            window_detected = True
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # Highlight window detection with green box
            cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), (0, 255, 0), 3)
            cv2.putText(annotated_frame, f"Window Detected ({conf:.2f})", 
                        (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                        0.7, (0, 255, 0), 2)

    # Display “spray” message if window found
    if window_detected:
        cv2.putText(annotated_frame, "Spray initiated on window!", (50, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)
        if not spray_announced:
            engine.say("Spray initiated on window")
            engine.runAndWait()
            spray_announced = True

    # Write and show
    out.write(annotated_frame)
    cv2.imshow("YOLO Window Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ===============================
# 4️⃣ Cleanup
# ===============================
cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Output saved as output_with_spray.mp4")
```
